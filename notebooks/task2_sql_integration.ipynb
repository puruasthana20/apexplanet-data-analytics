```python
import sqlite3
import pandas as pd

conn = sqlite3.connect("../data/customer_churn.db")

query = """
SELECT Contract,
AVG(Monthly_Charges) AS AvgCharge
FROM customer_churn
GROUP BY Contract;
"""

df = pd.read_sql(query, conn)

print(df)

conn.close()
```
