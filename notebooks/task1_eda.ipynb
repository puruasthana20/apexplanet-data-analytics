```python
import pandas as pd
import numpy as np

df = pd.read_csv("../data/raw/telco_churn.csv")

df.head()
```

```python
df.columns = (
    df.columns
      .str.strip()
      .str.replace(" ", "_")
)
```

```python
df = df.drop_duplicates()
```

```python
df["Total_Charges"] = pd.to_numeric(
    df["Total_Charges"],
    errors="coerce"
)
```

```python
print(df.columns.tolist())
```

```python
df.columns = df.columns.str.strip()
print(df.columns.tolist())
```

```python
df["Total_Charges"]
```

```python
df.isnull().sum()
```

```python
df["Total_Charges"].isnull().sum()
```

```python
df["Total_Charges"] = df["Total_Charges"].fillna(
    df["Total_Charges"].median()
)
```

```python
print(df.columns.tolist())
```

```python
sns.countplot(x="Churn_Label", data=df)
plt.show()
```

```python
sns.countplot(x="Gender", data=df)
plt.show()
```

```python
sns.countplot(
    x="Contract",
    data=df
)

plt.xticks(rotation=45)
plt.show()
```

```python
sns.histplot(
    df["Monthly_Charges"],
    kde=True
)

plt.show()
```

```python
df_encoded = df.copy()

for col in df_encoded.select_dtypes(
        include='object'):
    df_encoded[col] = (
        df_encoded[col]
        .astype('category')
        .cat.codes
    )
```

```python
plt.figure(figsize=(12,8))

sns.heatmap(
    df_encoded.corr(),
    cmap="coolwarm"
)

plt.show()
```

1. Month-to-month contracts have highest churn.

2. Customers with higher tenure churn less.

3. Electronic check users churn more.

4. Fiber optic customers show higher churn.

5. Long-term contracts improve retention.
