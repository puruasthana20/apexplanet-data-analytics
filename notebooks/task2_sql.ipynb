```python
query = """
SELECT COUNT(*) AS Total_Customers
FROM customer_churn;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT Contract,
AVG(Monthly_Charges) AS AvgCharge
FROM customer_churn
GROUP BY Contract;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT CustomerID,
Total_Charges
FROM customer_churn
ORDER BY Total_Charges DESC
LIMIT 10;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT CustomerID,
Total_Charges
FROM customer_churn
ORDER BY Total_Charges DESC
LIMIT 10;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT Payment_Method,
COUNT(*) AS Churned_Customers
FROM customer_churn
WHERE Churn_Label='Yes'
GROUP BY Payment_Method
ORDER BY Churned_Customers DESC;
"""

pd.read_sql(query, conn)
```
